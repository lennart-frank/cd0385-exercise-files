# Scatter Plot Of Features
Create a scatter plot of the four features against eachother to visualize the results from the correlation matrix
1. `sepal length (cm)` vs. `sepal width (cm)`
2. `sepal length (cm)` vs. `petal length (cm)`
3. `sepal length (cm)` vs. `petal width (cm)`
4. `sepal width (cm)` vs. `petal length (cm)`
5. `sepal width (cm)` vs. `petal width (cm)`
6. `petal length (cm)` vs. `petal width (cm)`

```python
fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(12, 8))
# Create the iris `data` dataset as a dataframe and name the columns with `feature_names`
df = pd.DataFrame(iris["data"], columns=iris["feature_names"])

# Include the target as well
df['target'] = iris["target"]

df.plot(ax=axes[0,0], kind="scatter", x="sepal length (cm)", y="sepal width (cm)")
df.plot(ax=axes[0,1], kind="scatter", x="sepal length (cm)", y="petal length (cm)")
df.plot(ax=axes[0,2], kind="scatter", x="sepal length (cm)", y="petal width (cm)")
df.plot(ax=axes[1,0], kind="scatter", x="sepal width (cm)", y="petal length (cm)")
df.plot(ax=axes[1,1], kind="scatter", x="sepal width (cm)", y="petal width (cm)")
df.plot(ax=axes[1,2], kind="scatter", x="petal length (cm)", y="petal width (cm)")
```

# Scatter Plot Of Features And Target
Create a scatter plot of the four features against the target
1. `sepal length (cm)`
2. `sepal width (cm)`
3. `petal length (cm)`
4. `petal width (cm)`

```python
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(10, 8))
df.plot(ax=axes[0, 0], kind="scatter", y="sepal length (cm)", x="target")
df.plot(ax=axes[0, 1], kind="scatter", y="sepal width (cm)", x="target")
df.plot(ax=axes[1, 0], kind="scatter", y="petal length (cm)", x="target")
df.plot(ax=axes[1, 1], kind="scatter", y="petal width (cm)", x="target")
```
